```python
# Data Cleaning and Preprocessing

# Import the pandas library for data manipulation and analysis
import pandas as pd
# Import the numpy library for numerical operations
import numpy as np

# Sample data
data = {
    'A': [1, 2, np.nan, 4],  # Define column 'A' with some NaN values
    'B': [5, np.nan, np.nan, 8],  # Define column 'B' with some NaN values
    'C': [9, 10, 11, 12]  # Define column 'C' with no NaN values
}
# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Handling missing values
# Forward fill missing values in the DataFrame
df.ffill(inplace=True)

# Removing outliers
# Calculate the first quartile (25th percentile) of each column
Q1 = df.quantile(0.25)
# Calculate the third quartile (75th percentile) of each column
Q3 = df.quantile(0.75)
# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1
# Remove rows with any values outside 1.5*IQR from Q1 and Q3
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Print the cleaned DataFrame
print(df)
```

```python
# Inferential Statistics
from scipy import stats

# Sample data
data = [1, 2, 3, 4, 5]

# Hypothesis testing (one-sample t-test)
t_stat, p_value = stats.ttest_1samp(data, 3)
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Confidence interval
confidence_interval = stats.t.interval(0.95, len(data)-1, loc=np.mean(data), scale=stats.sem(data))
print("95% Confidence Interval:", confidence_interval)
```

```python
# Pandas for Time Series
import pandas as pd
import numpy as np

# Creating a datetime index
dates = pd.date_range(start='1/1/2020', periods=100, freq='D')
data = np.random.randn(100)
ts = pd.Series(data, index=dates)

print(ts.head())
```

```python
# Simple Moving Average
rolling_mean = ts.rolling(window=7).mean()
print(rolling_mean.head(10))
```

```python
# Exponential Smoothing
ewma = ts.ewm(span=7, adjust=False).mean()
print(ewma.head(10))
```

```python
# Geopandas
import geopandas as gpd
from shapely.geometry import Point

# Create a GeoDataFrame
data = {
    'name': ['Restaurant A', 'Restaurant B', 'Restaurant C'],
    'cuisine': ['Italian', 'Chinese', 'Mexican'],
    'geometry': [Point(1, 1), Point(2, 2), Point(3, 3)]
}
gdf = gpd.GeoDataFrame(data, crs="EPSG:4326")

# Display the GeoDataFrame
print(gdf)
```

```python
# Mapping with Matplotlib and Folium
import matplotlib.pyplot as plt

# Plot the GeoDataFrame
gdf.plot(column='cuisine', cmap='viridis', legend=True)
plt.title('Restaurant Locations by Cuisine')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Folium usage
import folium
import geopandas as gpd
from shapely.geometry import Point

# Create a GeoDataFrame
data = {
    'name': ['Restaurant A', 'Restaurant B', 'Restaurant C'],
    'cuisine': ['Italian', 'Chinese', 'Mexican'],
    'geometry': [Point(1, 1), Point(2, 2), Point(3, 3)]
}
gdf = gpd.GeoDataFrame(data, crs="EPSG:4326")

# Create a base map
m = folium.Map(location=[2, 2], zoom_start=13)

# Add markers for each restaurant
for idx, row in gdf.iterrows():
    folium.Marker([row.geometry.y, row.geometry.x], popup=row['name']).add_to(m)

# Display the map
m
```

```python
# Geospatial Analysis Techniques with Spatial Joins
import geopandas as gpd
from shapely.geometry import Polygon

# Create a GeoDataFrame for neighborhoods
neighborhoods = gpd.GeoDataFrame({
    'name': ['Neighborhood 1', 'Neighborhood 2'],
    'geometry': [Polygon([(0, 0), (0, 2), (2, 2), (2, 0)]), Polygon([(2, 2), (2, 4), (4, 4), (4, 2)])]
}, crs="EPSG:4326")

# Create a GeoDataFrame for points or other geometries to be joined
# For example, let's create a GeoDataFrame with some points
points = gpd.GeoDataFrame({
    'id': [1, 2, 3],
    'geometry': [Point(1, 1), Point(3, 3), Point(5, 5)]
}, crs="EPSG:4326")

# Perform a spatial join
joined = gpd.sjoin(points, neighborhoods, how="inner")
print(joined)
```

```python
# Proximity Analysis:

from shapely.ops import nearest_points
import geopandas as gpd
from shapely.geometry import Point

# Create a GeoDataFrame for hospitals
hospitals = gpd.GeoDataFrame({
    'name': ['Hospital 1', 'Hospital 2'],
    'geometry': [Point(1.5, 1.5), Point(3.5, 3.5)]
}, crs="EPSG:4326")

# Find the nearest hospital for each restaurant
nearest_hospitals = []
for idx, row in gdf.iterrows():
    nearest = None
    min_distance = float('inf')
    for _, hospital in hospitals.iterrows():
        distance = row.geometry.distance(hospital.geometry)
        if distance < min_distance:
            min_distance = distance
            nearest = hospital.geometry
    nearest_hospital = hospitals[hospitals.geometry == nearest]['name'].values[0]
    nearest_hospitals.append(nearest_hospital)

gdf['nearest_hospital'] = nearest_hospitals
print(gdf)
```
